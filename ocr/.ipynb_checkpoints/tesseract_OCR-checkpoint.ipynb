**plt_imshow** : 이미지 출력해주는 함수

```python
#### plt_imshow ####
def plt_imshow(title='image', img=None, figsize=(8 ,5)):
    '''
    title - 이미지 제목. default=image.여러 이미지 표시 경우 제목 목록 전달 가능
    img - 이미지 또는 이미지 목록. 단일 이미지 또는 여러 이미지 처리 가능
    figsize - 이미지 표시 크기를 결정하는 숫자 쌍. default=(8,5)
    '''
    plt.figure(figsize=figsize)
 
    ## img가 이미지 목록인 경우
    if type(img) == list:
        if type(title) == list: titles = title  # title 목록 지정
        else: # title 목록 생성
            titles = []
            for i in range(len(img)):
                titles.append(title)
                
        # 모든 이미지 rgb 변환
        for i in range(len(img)):
            if len(img[i].shape) <= 2:
                rgbImg = cv2.cvtColor(img[i], cv2.COLOR_GRAY2RGB)
            else:
                rgbImg = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)
 
            plt.subplot(1, len(img), i + 1), plt.imshow(rgbImg)
            plt.title(titles[i])
            plt.xticks([]), plt.yticks([])
            
        plt.show()
    
    ## img가 단일 이미지인 경우
    else:  # rgb 변환
        if len(img.shape) < 3: rgbImg = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        else: rgbImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 
        plt.imshow(rgbImg)
        plt.title(title)
        plt.xticks([]), plt.yticks([])
        plt.show()
```

### 텍스트 윤곽선을 찾기 위해 gray, blurred, edged 찾기

```python
image_copy = image.copy()  # 기존 이미지 copy
image_copy = imutils.resize(image_copy, width=500)  # 이미지 폭을 500픽셀로 조정
ratio = image.shape[1] / float(image_copy.shape[1])  # 비율 계산
 
# 이미지를 grayscale로 변환하고 blur를 적용
# >> 모서리를 찾기 위한 이미지 연산
gray = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5,), 0)
edged = cv2.Canny(blurred, 75, 200)  # Canny():모서리 감지

# 모서리 찾는 과정 확인
plt_imshow(['gray', 'blurred', 'edged'], [gray, blurred, edged])
```

```python
#### contours를 찾아 크기순으로 정렬
# 1. findContours() : edged 이미지(복사본)의 윤곽을 찾음
#    RETR_EXTERNAL : 가장 외부 윤곽만 검색 
#    CHAIN_APPROX_SIMPLE : 윤곽점 줄임 >> 수직 및 수평 윤곽만 저장하며, 불필요한 정보 제거
# 2. imutils.grab_contours() : 윤곽 정보 추출
# 3. sorted() & cv2.contourArea : 윤곽 목록 내림차순 정렬 >> 가장 큰 영역의 윤곽이 목록의 첫번째 요소가 됨
cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)


receiptCnt = None

# 정렬된 contours를 반복문으로 수행하며 4개의 꼭지점을 갖는 도형을 검출
for c in cnts:
    peri = cv2.arcLength(c, True)  # 각 윤곽 둘레 계산(True-닫힌 도형)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)  # 윤곽을 근사(원래의 윤곽 간략화, 불필요한 정점 제거)
             # 0.02*peri = 근사 정확도 결정(작을수록 원래의 윤곽에 가까워짐, 클수록 단순한 도형), True-닫힌 도형

    # contours가 크기순으로 정렬되어 있기 때문에 제일 첫번째 사각형을 영수증 영역으로 판단하고 break
    if len(approx) == 4:
        receiptCnt = approx # 영수증의 윤곽 정보 저장
        break
 
 
# 만약 추출한 윤곽이 없을 경우 오류
if receiptCnt is None:
    raise Exception(("Could not find receipt outline."))
    
output = image.copy() # 이미지 복사본에 녹색 선 그리기
cv2.drawContours(output, [receiptCnt], -1, (0, 255, 0), 2)
# output에 윤곽 정보 전달
# -1(그리고자 하는 윤곽선 인덱스/-1은 모든 윤곽을 그림) 녹색(0,255,0)&두께 2픽셀 설정하고 그리기
plt_imshow("Receipt Outline", output)
```

```python
#### 1. 노이즈 제거
rm_noise = cv2.medianBlur(gray_image, ksize=3)

plt_imshow('REMOVE_NOISE', rm_noise)
```

```python
#### 2. Morphology 변환
kernel = np.ones((3,3), np.uint8)

# dilation = cv2.dilate(rm_noise, kernel, iterations=1)  # 팽창: 객체 외곽 픽셀 주변에 1(흰색) 추가
# erosion = cv2.erode(rm_noise, kernel, iterations=1)  # 침식: 객체 외곽 픽셀 주변에 0(검은색) 추가
open_ = cv2.morphologyEx(rm_noise, cv2.MORPH_OPEN, kernel)  # 침식 > 팽창: 이미지 상의 작은 잡티, 물체 등 제거 효과
# close_ = cv2.morphologyEx(rm_noise, cv2.MORPH_CLOSE, kernel)  # 팽창 > 침식: 전체적인 윤곽을 뚜렷하게 보이는 효과

plt_imshow(['Dilation','Erosion','Opening','Closing'],[dilation, erosion, open_, close_], (20,5))
```

```python
## 3. 대비 조절
def change_contrast(img, alpha=0.0):
    '''
    명암비(contrast)를 변환하는 함수
    [parameter]
        img: ndarray - 명암비를 조절할 대상 이미지
        alpha: float -  대비 조절 비율값. 기본: 0.0(원본) - 양수: 명암비를 높인다. 음수 - 명암비를 낮춘다. 
    [return]
        ndarray - 명암비가 변환된 이미지
    '''
    return np.clip((1.0 + alpha) * img - 128 * alpha, 0, 255).astype('uint8')

contrast = change_contrast(rm_noise, 1)
plt_imshow('contrast', contrast)
```

## 실시간 탐지

```python
## 웹캠 연동
cap = cv2.VideoCapture(0)

while True:
    # 한 Frame 읽기
    success, frame = cap.read()
    if not success:
        print('프레임을 읽지 못함')
        break
    
    # 이미지 전처리 (옵션)
    processed_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # grayscale
    processed_frame = cv2.medianBlur(processed_frame, ksize=3)  # noise 제거
    processed_frame = change_contrast(processed_frame, 1)  # 대비 높임

    # OCR 실행
    config = '--psm 6 outputbase digits'  # 숫자만 찾는 모드
    ocr_result = pytesseract.image_to_string(processed_frame, config=config)
    
    
    ############## 영역 추출 ##############
    ocr_result = re.findall("\d+", ocr_result)
    if ocr_result:
        for num in ocr_result:
            if len(num) == 4: text='worker id: ' + num  # 읽은 숫자의 길이가 4일 경우에만 출력
            else: text = 'no read'
    else: text = 'no read'
    ######################################
    
    
    
    # 인식된 텍스트를 화면에 출력
    font = cv2.FONT_HERSHEY_SIMPLEX
    frame_with_text = cv2.putText(frame, text, (10, 50), font, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    
    # 화면에 출력
    cv2.imshow('frame', frame_with_text)
    
    if cv2.waitKey(1) == 27: # esc 
        break
        
cv2.destroyAllWindows()
cap.release()
```
